```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Perceptron:
    def __init__(self, learning_rate=0.01, epochs=100):
        self.learning_rate = learning_rate
        self.epochs = epochs

    def fit(self, X, y):
        # Inicializar pesos y sesgo
        self.weights = np.zeros(X.shape[1])
        self.bias = 0
        
        # Entrenamiento del Perceptrón
        for _ in range(self.epochs):
            for i in range(X.shape[0]):
                # Calcular el producto punto entre pesos y características más el sesgo
                y_pred = np.dot(X[i], self.weights) + self.bias
                
                # Aplicar la función de activación (umbral) para obtener la salida
                y_pred = 1 if y_pred >= 0 else 0
                
                # Actualizar pesos y sesgo si hay un error en la clasificación
                if y_pred != y[i]:
                    update = self.learning_rate * (y[i] - y_pred)
                    self.weights += update * X[i]
                    self.bias += update
    
    def predict(self, X):
        # Predecir la clase para nuevas instancias de datos
        y_pred = np.dot(X, self.weights) + self.bias
        return np.where(y_pred >= 0, 1, 0)
```

```python
# Datos de ejemplo
X = np.array([[2, 3], [1, 2], [5, 2], [3, 1]])
y = np.array([0, 0, 1, 1])
```

```python
# Crear instancia del Perceptrón y entrenarlo
perceptron = Perceptron()
perceptron.fit(X, y)
```

```python
# Generar gráfico de dispersión de los datos
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k', label='Datos')

# Crear hiperplano de separación
x_vals = np.linspace(0, 5, 10)
y_vals = (-perceptron.weights[0] * x_vals - perceptron.bias) / perceptron.weights[1]
plt.plot(x_vals, y_vals, 'k--', label='Hiperplano de separación')

plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Perceptrón: Clasificación de Datos')
plt.legend()
plt.grid(True)
plt.show()
```
