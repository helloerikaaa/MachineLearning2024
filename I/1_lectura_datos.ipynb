### Datos Tabulares
Los datos tabulares se refieren a conjuntos de datos organizados en forma de tabla, donde las filas representan observaciones individuales y las columnas representan variables o características específicas de esas observaciones. Cada celda en la tabla contiene un valor único correspondiente a una combinación específica de fila y columna.

```python
# Eliminar Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lectura de Datos Tabulares
datos = pd.read_csv("../data/netflix.csv")
```

```python
# Mostrar primeros 5 datos
datos.head(5)
```

```python
# Mostrar últimos 5 datos
datos.tail(5)
```

```python
# Información del archivo CSV
datos.info()
```

```python
# Describir los datos
datos.describe()
```

```python
# Seleccionar una columna
pais = datos['country']
pais
```

```python
cols = datos[['title', 'director']]
cols
```

```python
pelis_filtradas = datos[datos['release_year'] < 2019]
pelis_filtradas
```

```python
# Valores Faltantes
print(datos.isnull().sum())
```

```python
# Ordenar el DataFrame por una columna específica
datos_ordenados = datos.sort_values(by='release_year')
datos_ordenados
```

```python
# Eliminar una columna específica
datos_sin_id = datos.drop('show_id', axis=1)
datos_sin_id
```

```python
# Eliminar filas con índices específicos
datos_sin_filas = datos.drop([0, 1, 2])
datos_sin_filas
```

```python
# Aplicar una función lambda a cada elemento de una columna
datos['new_year'] = datos['release_year'].apply(lambda x: x - 10)
datos
```

```python
# Distribución de datos
plt.figure(figsize=(12, 8))
plt.hist(datos['release_year'], bins=30)
plt.title('Distribución de variable numérica')
plt.xlabel('Valor numérico')
plt.ylabel('Frecuencia')
plt.show()
```

### Imágenes Digitales
Las imágenes digitales son representaciones visuales de datos almacenadas en forma de píxeles, donde cada píxel contiene información sobre el color y la intensidad luminosa en una ubicación específica de la imagen. Las imágenes digitales pueden ser fotografías, ilustraciones, gráficos, etc.

```python
from PIL import Image
```

```python
# Abrir una imagen
imagen = Image.open('../data/perrito.jpeg')
```

```python
# Mostrar una imagen afuera del Notebook
imagen.show()
```

```python
# Mostrar una imagen adentro del Notebook
plt.imshow(imagen)
plt.axis('off')
```

```python
# Cambiar el tamaño de la imagen
imagen_redimensionada = imagen.resize((200, 400))
```

```python
plt.imshow(imagen_redimensionada)
plt.axis('off')
```

```python
# Convertir la imagen a escala de grises
imagen_escala_grises = imagen.convert('L')
```

```python
plt.imshow(imagen_escala_grises, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
# Región de Interés
roi = (10, 50, 30, 80)
imagen_recortada = imagen.crop(roi)
```

```python
plt.imshow(imagen_recortada)
plt.axis('off')
```
