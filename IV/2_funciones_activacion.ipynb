```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definicin de las funciones de activacin
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Rango de valores para x
x = np.linspace(-5, 5, 100)

# Calcular los valores de y para cada funcin de activacin
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

lista_sigmoid = [sigmoid(xi) for xi in x]
lista_relu = [relu(xi) for xi in x]
lista_tanh = [tanh(xi) for xi in x]

df = pd.DataFrame({
    'x': x,
    'y_sigmoid': lista_sigmoid,
    'y_relu': lista_relu,
    'y_tanh': lista_tanh
})

print(df.head())
print(df.tail())

# Graficar las funciones de activacin
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, label='Sigmoid')
plt.title('Funcin Sigmoide')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(x, y_relu, label='ReLU', color='orange')
plt.title('Funcin ReLU')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, label='Tanh', color='green')
plt.title('Funcin Tangente Hiperblica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
